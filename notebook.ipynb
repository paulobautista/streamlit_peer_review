```python
import boto3

bucket_name = 'streamlit-images'
region = 'ap-southeast-1'  # e.g., 'us-east-1'

s3_client = boto3.client('s3', region_name=region)

paginator = s3_client.get_paginator('list_objects_v2')
pages = paginator.paginate(Bucket=bucket_name)

with open('image_urls.csv', 'w') as file:
    for page in pages:
        for obj in page.get('Contents', []):
            key = obj['Key']
            url = f'https://{bucket_name}.s3.{region}.amazonaws.com/{key}'
            file.write(f"{key},{url}\n")
```

```python
import os
import uuid
import pandas as pd

def create_document_entries(images_dir: str, markdown_dir: str, output_csv: str):
    files_with_no_md = 0
    """
    Create document entries from image and markdown files.

    Args:
        images_dir: Path to directory containing image files
        markdown_dir: Path to directory containing markdown files
        output_csv: Path where to save the CSV file
    """
    # Get a list of all image files
    image_files = [f for f in os.listdir(images_dir) if f.endswith(('.png', '.jpg', '.jpeg'))]

    # Initialize a list to hold document data
    documents = []

    for index, image_file in enumerate(image_files, start=1):
        index = index + 198
        # Generate a UUID
        document_uuid = str(uuid.uuid4())

        # Extract the base filename without extension
        base_filename = os.path.splitext(image_file)[0]

        # Construct the full paths
        image_path = os.path.join(images_dir, image_file)
        markdown_filename = f"{base_filename}.md"
        markdown_path = os.path.join(markdown_dir, markdown_filename)

        # Read the markdown content if it exists
        if os.path.exists(markdown_path):
            with open(markdown_path, 'r', encoding='utf-8') as md_file:
                markdown_content = md_file.read()
        else:
            markdown_content = ''
            print(f"Warning: Markdown file not found for {image_file}")
            files_with_no_md += 1


        # Create document entry
        documents.append({
            'id': document_uuid,
            'doc_index': index,
            'image_path': image_path,
            'markdown': markdown_content
        })

    print(files_with_no_md)

    # Create a DataFrame and save to CSV
    df = pd.DataFrame(documents)
    df[['doc_index', 'image_path', 'markdown']].to_csv(output_csv, index=False)

    print(f"Created {len(documents)} document entries")
    print(f"Saved to {output_csv}")
```
